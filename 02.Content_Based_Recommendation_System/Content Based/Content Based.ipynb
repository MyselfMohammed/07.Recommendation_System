```python
# Created a Function - 
def Whole_Users_Vector(Dataset, Genre_Vector_Transpose):
    # Create a Vector List for Whole Users with Movie Vector
    Whole_Users_Vector_List = []
    
    Total_Number_Of_Users = range(1,944)
    Total_Number_Of_Genre = range(1,19)
    Total_Number_Of_Iteration = 0
    
    for Each_User in Total_Number_Of_Users :
        
        Users = Dataset[Dataset['user_id']==Each_User][['movie_id','rating']]
        
        Users_Movie_Id_List = []
    
        for Each_Movie_Id in Users['movie_id']:
            User_Movie_Id = Genre_Vector_Transpose[Each_Movie_Id].values
            Users_Movie_Id_List.append(User_Movie_Id)
        
        # Converting the DataFrame to dataframe
        Users_Movie_Genre_Table = pd.DataFrame(Users_Movie_Id_List, columns = Total_Number_Of_Genre)
            
        # Create Single Vector for One User by Summing and Taking Average of Each Genre Column
        Single_User_Vector_List = []
        
        for Each_Genre in Total_Number_Of_Genre:
            Summed_Each_Genre_Column_Value = sum(Users_Movie_Genre_Table[Each_Genre])
            #print(f"Summed Each Genre Column - {Each_Genre} Value : {Summed_Each_Genre_Column_Value}")
            
            Each_Genre_Vector_Value = (Summed_Each_Genre_Column_Value / len(Users))
            Single_User_Vector_List.append(Each_Genre_Vector_Value)
            #print(f"Genre - {Each_Genre} Vector Value = ({Summed_Each_Genre_Column_Value})/ Total_Number_Of_Rows_In_Users ({len(Users)}) = {Each_Genre_Vector_Value:.2f}", "\n")

            Total_Number_Of_Iteration = Total_Number_Of_Iteration + 1
            
        Whole_Users_Vector_List.append(Single_User_Vector_List)
    
    print(f"Total Number Of Iteration in Function - Whole_Users_Vector : {Total_Number_Of_Iteration}")
    return Whole_Users_Vector_List
```

```python
# Called the Function - Whole_Users_Vector
Vectors = Whole_Users_Vector(Dataset, Genre_Vector_Transpose)
```

```python
# Converting into Dataset
Users_Vector = pd.DataFrame(Vectors)
Users_Vector.head()
```
